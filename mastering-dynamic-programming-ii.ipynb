# Best Time to Buy and Sell Stock IV

```python
def maxProfitBottomUp(k, prices):
        
    if len(prices) == 0 or k == 0:
        return 0

    # Initialize 2 arrays with value of zeros,
    # to store profits for current transaction (txn) and previous transaction (txn-1).
    even_array = [0 for _ in prices]
    odd_array = [0 for _ in prices]

    # Iterate through the transactions, starting from row 1.
    for txn in range(1, k+1):

        p_max = float("-inf")       # Initialize p_max with negative infinity.      

        if txn % 2 == 1:
            curr_txn = odd_array    # We first start from row 1, so current transaction is on odd_array.
            prev_txn = even_array  
        else:
            curr_txn = even_array
            prev_txn = odd_array             

        # Iterate through the prices of each day, starting from column 1.
        for day in range(1, len(prices)):                      

            p_max = max(p_max, prev_txn[day-1] - prices[day-1])  
            
            scenario1 = curr_txn[day-1]
            scenario2 = prices[day] + p_max
            
            curr_txn[day] = max(scenario1, scenario2)
    
    # Return the last element in the array.
    return even_array[-1] if k % 2 == 0 else odd_array[-1]
```
